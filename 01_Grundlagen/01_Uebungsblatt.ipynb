### Aufgabe 1: Temperatur in Davos


#### a) Daten importieren

In der Datei `DavosTemperatur.txt` sind drei Spalten, die erste gibt das Kürzel der Wetterstation an, die zweite zeigt eine Zahl welche aus dem Datum und der Stunde besteht. Die dritte Spalte gibt die gemessene Temperatur wieder.
Importiere die dritte Spalte der Datei `DavosTemperatur.txt` in ein Array namens `temperature`. Gib den Shape des Arrays und die ersten 5 Elemente aus.


```python
# Import der dritten Spalte in das Array "temperature"
temperature = np.loadtxt("DavosTemperatur.txt", delimiter=";", skiprows=1, usecols=2)

# Ausgabe des Shapes des Arrays und der ersten 5 Elemente
print("Shape des Arrays:", temperature.shape)
print("Die ersten 5 Elemente:", temperature[:5])
```

#### b) Daten darstellen

- Stelle die Temperaturen grafisch dar.
- Beschrifte die Achsen und gib dem Plot einen aussagekräftigen Titel.
- Schränke den Bereich der x-Achse (Zeit) auf die ersten 48 Stunden ein. Benutze dazu die Funktion [`plt.xlim`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.xlim.html).


```python
plt.plot(temperature)
plt.xlabel("Zeit (Stunden)")
plt.ylabel("Temperatur (°C)")
plt.title("Gemessene Temperaturen in Davos")
plt.xlim(0, 48)
plt.show()
```

### Aufgabe 3: Bakterien


#### a) Daten importieren

Importiere die Daten in der Datei `bacteria_nutrient.csv` in ein Array namens `bac_data`. Gib den Shape des Arrays und die ersten 5 Zeilen aus.

Wie viele Beobachtungen sind es und welche Variablen sind in den Spalten zu finden? Übertrage die Spalten in entsprechend benannte 1D-Arrays.


```python
bac_data = np.loadtxt("bacteria_nutrient.csv", delimiter=",",skiprows=1)
print("Shape des Arrays:", bac_data.shape)
print("Die ersten 5 Elemente:")
print(bac_data[:5])
```

```python
# Lese die erste Zeile der Datei
with open('bacteria_nutrient.csv', 'r') as f:
    header = f.readline()

# Extrahiere die Namen der Variablen
var_names = header.strip().split(',')

# Gib die Anzahl der Variablen aus
print("Anzahl der Variablen:", len(var_names))

# Drucke die Namen der Variablen
print("Variablen:", var_names)

zeit = bac_data[:,0]
bacteria = bac_data[:,1]
nutrient = bac_data[:,2]
print("Zeit:",zeit)
print("Bakterien",bacteria)
print("Nährstoff",nutrient)
```

#### b) Bakterienanzahl darstellen

- Stelle den Verlauf der Bakterienanzahl und Nahrungsmenge über die Zeit dar. Erzeuge dafür mit [`plt.subplots`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html) zwei übereinanderliegende Koordinatensysteme.
- Beschrifte die Achsen und geben dem Plot einen Aussagekräftigen Titel (diesmal mit [`plt.suptitle`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.suptitle.html)).
- Mit [`plt.xlim`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.xlim.html) kannst du wieder die x-Achse so wählen, dass der Graph durchgängig von einer zur anderen Seite geht.
- Füge dem Graphen einen Gitternetz hinzu. Benutze dafür [`plt.grid`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.grid.html).


```python
# Zwei Koordinatensysteme übereinanderlegen
fig, (ax1, ax2) = plt.subplots(2,1) #sharex Teilt die gleichen X Achsen 

# Bakterienanzahl plotten
ax1.plot(zeit, bacteria, 'r-', label='Bakterienanzahl')
ax1.set_ylabel('Bakterienanzahl')
ax1.legend(loc='upper left')

# Nahrungsmenge plotten
ax2.plot(zeit, nutrient, 'b-', label='Nährstoffmenge')
ax2.set_xlabel('Zeit')
ax2.set_ylabel('Menge')
ax2.legend(loc='upper left')

# Achsenbeschriftung und Titel hinzufügen
plt.suptitle('Bakterienwachstum in Abhängigkeit von der Nahrungsmenge')
ax2.set_xlim(left=zeit[0], right=zeit[-1])

# Gitternetz Hinzufügen 
ax1.grid(True)
ax2.grid(True)
plt.show()
```

#### c) Nährmittelmenge darstellen

Stelle analog zur Bakterienanzahl den Verlauf der Nährmittelmenge dar.


```python
ax2 = fig.add_subplot(2, 1, 2)

# Plotten der Nährmittelmenge gegen die Zeit
ax2.plot(zeit, nutrient, color='green', marker='o')

# Achsenbeschriftungen und Titel hinzufügen
ax2.set_xlabel('Zeit (h)')
ax2.set_ylabel('Nährmittelmenge (mg)')
ax2.set_title('Verlauf der Nährmittelmenge')

# Setzen der x-Achsen-Grenzen, des Gitternetzes und des Layouts
ax2.set_xlim([0, zeit[-1]])
ax2.grid()
fig.tight_layout()
plt.show()
```

#### d) Bakterienanzahl und Nährstoffmenge zusammen darstellen

- Stelle den Verlauf der Bakterienanzahl und der Nährstoffmenge zusammen über die Zeit dar.
- Beide sollten sich eine x-Achse teilen und übereinander platziert sein. Dies kannst du über [`plt.subplots`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html) erreichen.
- Setze einen einzigen Titel für das ganze Diagramm mit [`plt.suptitle`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.suptitle.html)


```python
fig, ax = plt.subplots()
plt.suptitle("Bakterienanzahl und Nährstoffmenge")
ax.set_xlabel('Zeit (h)')
ax.set_ylabel("Weiss nicht")
ax.plot(zeit, bacteria,  color='green', label="Bakterienanzahl")
ax.plot(zeit, nutrient,  color='red', label="Nährstoffmenge")
ax.legend()
ax.plot()
```

#### e) Hinweis

Zwei Graphen unterschiedlicher Einheiten (wie Bakterienanzahl und Nährmittelmenge) sollten wenn möglich nicht im gleichen Diagramm (Koordinatensystem) dargestellt werden. Die Variante mit den zwei Subplots wie oben ist vorzuziehen.


### Aufgabe 4: Darstellung von 2D-Arrays / Matrizen


Versuche folgenden Plot nachzustellen:




```python
# Fixing random state for reproducibility
np.random.seed(19680801)


X = 10*np.random.rand(5, 3)

fig, ax = plt.subplots()
ax.imshow(X)


def format_coord(x, y):
    col = round(x)
    row = round(y)
    nrows, ncols = X.shape
    if 0 <= col < ncols and 0 <= row < nrows:
        z = X[row, col]
        return f'x={x:1.14f}, y={y:1.14f}, z={z:1.14f}'
    else:
        return f'x={x:1.14f}, y={y:1.14f}'


ax.format_coord = format_coord
plt.show()
```

### Aufgabe 1: Simulieren von Noise


#### a) Erzeugen von Daten

Wir betrachten einen Zeitraum $0$ bis $5$, benutze hierfür [`np.linspace`](https://numpy.org/doc/stable/reference/generated/numpy.linspace.html). Es wird folgende Funktion für die Daten erzeugung verwendet $y = 2 \sin(2 \pi t) + 10$.


```python
b = np.linspace(0,5)
t = np.linspace(0, 5, num=1000) 
#num macht, dass die kurven gleichmässiger sind
y1=  2 * np.sin(2 * np.pi * b) + 10
y2 = 2 * np.sin(2 * np.pi * t) + 10
y1

plt.plot(t,y2)

plt.grid()
plt.show
```

#### b) Hinzufügen von Noise

- Erzeuge normalverteilte Messfehler (Zufallszahlen) in einem Array namens `noise`. Diese sollen um $0$ mit einer Varianz von $1$ verteilt sein. Speichere 
- Addiere Messfehler zu `y` und speichere die simulierten Daten in `noisy`. 

```python
rng = np.random.default_rng(seed=1) #seed generiert immer wieder die gleichen Zahlen
noisy = rng.normal(y1) #0 ist der median 
noisy
```

#### c) Darstellen der Funktion mit und ohne Messfehler

- Stelle `y` und `noisy` über die Zeit im gleichen Diagramm dar.
- Die Theorie `y` soll mit einer verbunenen Linie dargestellt werden, die Messwerte `noisy` nur als Punkte.
- Füge mit [`plt.legend`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.htmlml) eine Legende hinzu.
- Schreibe Achsen an.

```python
plt.plot(noisy,"-o")
plt.plot(y1, color="green",marker= "o")
plt.title("Funktion mit und ohne Messfehler")
plt.grid()
plt.show()
```

#### d) Abspeichern der Daten

Füge die drei Variablen Zeit, `y` und `noisy` in einer CSV-Datei `Noisy.csv` ab. Für das Zusammenfügen gibt es [`np.column_stack`](https://numpy.org/doc/stable/reference/generated/numpy.column_stack.html) und zum rausschreiben [`np.savetxt`](https://numpy.org/doc/stable/reference/generated/numpy.savetxt.html). Füge dem CSV-File eine aussagekräftige Zeilenbeschriftung hinzu.


```python
data = np.column_stack((noisy,y1,b))
np.savetxt("Noisy.csv", data, delimiter=";", header="Noisy, Funktionswerte, Zeit")
```
