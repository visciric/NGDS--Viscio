### Bisektionsverfahren

**Bei FBisektion/Newton  Verfahren kann man immer nur ``EINE`` Nullstelle berechnen !!!**

Das einfachste Nullstellenverfahren beruht auf dem [Zwischenwertsatz](https://de.wikipedia.org/wiki/Zwischenwertsatz) der Analysis. Dieser besagt, dass eine stetige Funktion auf einem Intervall $[a,b]$ sicher alle Werte zwischen $f(a)$ und $f(b)$ annimmt. Haben $f(a)$ und $f(b)$ unterschiedliche Vorzeichen, so muss also eine Nullstelle dazwischen liegen. Dies nutzt man, um die Nullstelle durch iteratives Teilen des Intervalls einzugrenzen.

### Newton-Verfahren

Das Newton-Verfahren nutzt die Linearisierung der Funktion, deren Nullstelle zu berechnen ist. Ausgehend von einer Schätzung der Nullstelle liefert die Nullstelle der Linearisierung eine neue Näherung der gesuchten Nullstelle. Diese Iteration konvergiert in der Regel schnell.

```python
# Newton Verfahren

def f(x):
    return x**3 + 3 * x**2 - x - 8


def f_prime(x):
    return 3 * x**2 + 6 * x - 1


def newton(x0, n):
    counter = 0
    x = x0
    while counter < n:
        x_new = x - (f(x) / f_prime(x))
        x = x_new
        counter += 1
    return x_new


print(newton(1, 4))
```

Beim Bisektionsverfahren ist es Wichtig, dass man sich zu erst überlegt was der Alghorithmus machen soll 

**Funktion Schreben -> return definieren** 

Kleiner Tipp man kann das Produkt berechnen und wenn es Positiv ist, nimmmt man den A wert wenn es negativ ist nimmmt man den B wert 
 

```python
#Newton Verfahren
def f(x):
    return x**3 + 3 * x**2 - x - 8


def bisection(
    f, a0, b0, n
):  # Funktion , a0 -> erstes Intervall b0 -> zweites Intervall
    a, b = a0, b0
    ya = f(a)

    steps = 0

    while steps < n:
        # als erstes Mitte bestimmen
        x = (b + a) / 2  # als erstes Mitte bestimmen
        y = f(x)

        # Produkt anschauen und wenn positiv dann nimmt es den x wert b/x und wenn negativ dann  a/x

        # hier passiert der Wechsel
        if ya * y < 0:
            b = x

        else:
            a = x
            ya = y
        steps += 1

    return x


bisection(f, 1, 2, 4)
```
