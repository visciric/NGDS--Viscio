### Aufgabe 1

Betrachten wir mal die Funktion

$$ g(x, y) = x \cdot e^{- x^2 - y^2} + 0.2y$$
<!-- $$ g(x, y) = \left(x^2 + 1.5 y^3\right) e^{- x^2 - y^2} $$ -->

im Wertebereich

$$x, y \in[-2,2]$$

**1.** Stelle die Höhenlinien dieser Funktion dar.

```python
import numpy as np
import matplotlib.pyplot as plt


def f(x, y):
    return x * np.exp(-(x**2) - y**2) + 0.2 * y


n = 100

x = np.linspace(-2, 2, n)
y = np.linspace(-2, 2, n)

X, Y = np.meshgrid(x, y)

Z = f(X, Y)
```

```python
# Aufgabe 1.1

plt.contour(X, Y, Z, levels=100, cmap="jet")
plt.colorbar(label="z")
CS = plt.contour(X, Y, Z, colors="black")
plt.clabel(CS, inline=1, fontsize=10)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

**2.** Stelle Isokonturlinien für a) konstante $x$ und b) konstante $y$ dar (mit korrekten Achsenbeschriftungen).

```python
# Aufgabe 1.2 a)

plt.contour(Y, Z, X, levels=100, cmap="jet")
plt.colorbar(label="X")
CS = plt.contour(Y, Z, X, colors="black")
plt.clabel(CS, inline=1, fontsize=10)
plt.xlabel("y")
plt.ylabel("z")
plt.show()
```

```python
# Aufgabe 1.2 b)

plt.contour(X, Z, Y, levels=100, cmap="jet")
plt.colorbar(label="Y")
CS = plt.contour(X, Z, Y, colors="black")
plt.clabel(CS, inline=1, fontsize=10)
plt.xlabel("x")
plt.ylabel("z")
plt.show()
```

**3.** Erstelle einen Höhenliniendiagramm mit gefüllten Flächen.

Nutze dazu die Funktion [`matplotlib.pyplot.contourf`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.contourf.html).

```python
# Aufgabe 1.3
plt.contourf(X, Y, Z, levels=100, cmap="jet")
plt.colorbar(label="z")
CS = plt.contour(X,Y,Z, colors='black')
plt.clabel(CS, inline=1, fontsize=10)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

**4.** Zeige alle Graphen in einem Plot

```python
# Aufgabe 1.4

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(8, 8), sharex=True)

ax1.contour(X, Y, Z, levels=25, cmap="jet")
ax1.clabel(ax1.contour(X, Y, Z, colors="black"), inline=1, fontsize=10)


ax2.contour(X, Z, Y, levels=25, cmap="jet")
ax2.clabel(ax2.contour(X, Z, Y, colors="black"), inline=1, fontsize=10)


ax3.contourf(X, Y, Z, levels=25, cmap="jet")
ax3.clabel(ax3.contour(X, Y, Z, colors="black"), inline=1, fontsize=10)

ax4.contourf(X, Z, Y, levels=25, cmap="jet")
ax4.clabel(ax4.contour(X, Z, Y, colors="black"), inline=1, fontsize=10)

plt.tight_layout()
plt.show()
```
