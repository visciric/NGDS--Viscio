### Funktionen mit mehreren Variablen

Viele Grössen in naturwissenschaftlichen und technischen Kontexten hängen von mehreren Variablen ab:

- Die Temperatur in einem Raum hängt von der 3D-Position $(x,y,z)$ ab.
- Das Wachstum einer Zellkutur hängt nicht nur von der Zeit $t$, sondern auch von der Menge der vorhandenen Nährstoffe, vom pH-Wert der Umgebung ab und wohl noch von anderen Variablen.

So gibt es auch in der Mathematik Funktionen, die von mehreren Variablen abhängen. Häufig heissen die unabhängigen Variablen $x$, $y$ und $z$, können aber auch mit einem Index durchgezählt ($x_1$, $x_2$, ...) oder in einem Vektor $\vec x \in \mathbb{R}^n$ zusammengepackt werden. Ein Beispiel:

$$ f(x,y) = x^2 - y^2 $$

### Grundsätzliches Vorgehen

Es sollen Linien konstanter Funktionswerte $z$ in der $x$-$y$-Ebene dargestellt werden.

1. Die Funktionsgleichung $z=f(x,y)$ entweder nach $x$ oder nach $y$ auflösen, je nach dem was einfacher ist. Falls es mehrere Lösungen gibt, z.B. beim Wurzelziehen, dann müssen alle diese Lösungen dargestellt werden. 
2. Bereich in $x$ (bzw. $y$) bestimmen, in dem die Funktion dargestellt werden soll. Dieser hängt stark vom Kontext ab. Array mit $x$-Werten (bzw. $y$-Werten) definieren.
3. Zugehörige $y$-Werte (bzw. $x$-Werte) für eine Serie von $z$-Werten berechnen und darstellen.

Für die Werte  $x\in[0,3]$ soll die Höhenlinie für $z=1$ dargestellt werden. Probiere dies zuerst mit Stift und Papier. Nutze hierfür eine Wertetabelle mit 4 $x$-Werten, die Du hier mithilfe von NumPy oder einer List-Comprehension ausrechnest.

### Teil 2: Mit Matplotlib

**2. a)** Nehmen wir mal $x\in[0,3]$ und stellen die Höhenlinie zu $z=1$ dar:

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def y_z(z, x):
    return z - 2 * x**2
```

```python
# Aufgabe 2a:

x = np.linspace(0, 3)
y = y_z(1, x)
plt.plot(x, y, label="z = 1")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

**2. b)** Nun machen wir das gleiche für mehrere $z$-Werte, sagen wir $z\in\{-5, -3, 0, 3, 5\}$ (`levels`) und stellen dies in einem Plot grafisch da.

```python
# Aufgabe 2b:

levels = (-5, -3, 0, 3, 5)
x = np.linspace(0, 3)
for z in levels:
    y = y_z(z, x)
    plt.plot(x, y, label=f"z = {z}")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

### Teil 3: Erstes Beispiel mit "Contour"

Meist kennt man eher den Bereich in der $x$-$y$-Ebene, für den man sich interessiert. Oder man kann die Funktion nicht von Hand nach einer der Variablen auflösen. Oder man hat Daten statt eine analytische Funktion gegeben. Zudem könnte es schwierig sein, die $z$-Werte (Levels) selbst zu wählen. In all diesen Fällen hilft einem die Funktion [`matplotlib.pyplot.contour`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.contour.html). 

Schau dir das erste Beispiel in der [Contour Demo](https://matplotlib.org/stable/gallery/images_contours_and_fields/contour_demo.html) an und versuche dann die folgenden Aufgaben zu lösen. 

**3. a)** Stelle die Funktion von oben im Bereich $x\in[0,3]$ und $y\in[-6,6]$ mithilfe von `plt.contour` dar.

```python
# Aufgabe 3a:

x = np.linspace(0, 3)
y = np.linspace(-6, 6)
X, Y = np.meshgrid(x, y)
Z = 2 * X**2 + Y
plt.contour(X, Y, Z)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

**3. b)** Finde heraus, welche $z$-Werte den dargestellen Höhenlinien entsprechen, indem Du mit [`matplotlib.pyplot.colorbar`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.colorbar.html) eine Farbwerteskala hinzufügst.

```python
# Aufgabe 3b:

x = np.linspace(0, 3)
y = np.linspace(-6, 6)
X, Y = np.meshgrid(x, y)
Z = 2 * X**2 + Y
plt.contour(X, Y, Z)
plt.xlabel("x")
plt.ylabel("y")
plt.colorbar(label="z")
plt.show()
```

**3. c)** Nutze die `levels` aus Aufgabe 2. b) in `contour` und stelle die dazu passenden Höhenlinien dar.

In `plt.contour` können die gewünschten Levels mit der Option `levels` angegeben werden. Zudem kann man mit der Methode `clabel()` noch die Levels im Diagramm anschreiben, anstatt mit der Colorbar zu arbeiten wie oben, siehe Beispiel in [Contour Demo](https://matplotlib.org/stable/gallery/images_contours_and_fields/contour_demo.html).

```python
# Aufgabe 3c:

levels = (-5, -3, 0, 3, 5)
x = np.linspace(0, 3)
y = np.linspace(-6, 6)
X, Y = np.meshgrid(x, y)
Z = 2 * X**2 + Y
cs = plt.contour(X, Y, Z, levels=levels)
cs.clabel(inline=True, fontsize=10)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
