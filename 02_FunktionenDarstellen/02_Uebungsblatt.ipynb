### Aufgabe 1

Betrachten wir mal die Funktion

$$ g(x, y) = x \cdot e^{- x^2 - y^2} + 0.2y$$
<!-- $$ g(x, y) = \left(x^2 + 1.5 y^3\right) e^{- x^2 - y^2} $$ -->

im Wertebereich

$$x, y \in[-2,2]$$

**1.** Stelle die Höhenlinien dieser Funktion dar.

**Wenn man Funktionen mit mehren Variablen hat muss man immer z= x,y haben** 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Aufgabe 1.1

def gr_1(x,y):
    return x * np.exp ** (-x ** 2 - y ** 2) + 0.2 * y

n = 100

x = np.linspace(-2, 2, n)
y = np.linspace(-2, 2, n)
X , Y = np.meshgrid(x,y)
```

**2.** Stelle Isokonturlinien für a) konstante $x$ und b) konstante $y$ dar (mit korrekten Achsenbeschriftungen).

```python
# Aufgabe 1.2 a)

X , Y = np.meshgrid(x,y)
Z = X * np.exp(-X **2 - Y ** 2) + 0.2 * Y 

plt.contour(X, Y, Z, levels=100, cmap="jet")
plt.xlabel("x")
plt.ylabel("y")

plt.colorbar(label = "z")

```

```python
# Aufgabe 1.2 b)

# von der Seite betrachtet 

X , Y = np.meshgrid(x,y)
Z = X * np.exp(-X **2 - Y ** 2) + 0.2 * Y 

plt.contour(X, Z, Y)
plt.xlabel("x")
plt.ylabel("y")
plt.colorbar(label = "z")
```

**3.** Erstelle einen Höhenliniendiagramm mit gefüllten Flächen.

Nutze dazu die Funktion [`matplotlib.pyplot.contourf`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.contourf.html).

```python
# Aufgabe 1.3
X , Y = np.meshgrid(x,y)
Z = X * np.exp(-X **2 - Y ** 2) + 0.2 * Y 

plt.contourf(X, Z, Y)
plt.xlabel("x")
plt.ylabel("y")
plt.colorbar(label = "z")
                                         


```

**4.** Zeige alle Graphen in einem Plot

```python
# Aufgabe 1.4
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(8, 8))

# Constant z contours on the lower left
cs = ax3.contour(X, Y, Z)
cs.clabel(inline=True, fontsize=9)
ax3.set_title("constant z")
ax3.set_xlabel("x")
ax3.set_ylabel("y")

# Constant x contours on the lower right
cs = ax4.contour(Z, Y, X)
cs.clabel(inline=True, fontsize=9)
ax4.set_title("constant x")
ax4.set_xlabel("z")
# ax4.set_ylabel('y')

# Constant y contours on the upper left
cs = ax1.contour(X, Z, Y)
cs.clabel(inline=True, fontsize=9)
ax1.set_title("constant y")
# ax1.set_xlabel('x')
ax1.set_ylabel("z")

# Empty subplot on the upper right
ax2.set_axis_off()
plt.show()
```

**2.** Erstelle für die eingelesenen Daten alle Darstellungen wie in Aufgabe 1.

Nutze [`numpy.linspace`](https://numpy.org/doc/stable/reference/generated/numpy.linspace.html) und [`numpy.meshgrid`](https://numpy.org/doc/stable/reference/generated/numpy.meshgrid.html), um die zugehören $x$- und $y$-Matrizen zu erzeugen. Plotte die Contours und packe alle Plots in ein einziges Diagramm wie in Aufgabe 1.

```python
# Aufgabe 2.2

x = np.linspace(-2, 2, Z.shape[0])
y = np.linspace(-2, 2, Z.shape[1])
X,Y = np.meshgrid(x,y)

"""
shape [0], nimmt die erste spalte und shape[1], die eine spalte 
wird dann transponiert

mit linspace werden Werte die zwischen -2 und 2 liegen gewählt und rechtmässig verteilt. 

meshgrid rechnet dann die Tabelle aus 
    --> die ausgerechneten Werte ergeben Punkte für eine Funktion 
"""

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(8, 8), constrained_layout=True)

ax1.contour(Y, Z, X)
ax1.set_xlabel("y")
ax1.set_ylabel("z")
ax1.set_title("x constant")

ax2.contour(X, Z, Y)
ax2.set_xlabel("x")
ax2.set_ylabel("z")
ax2.set_title("y constant")

ax3.contour(X, Y, Z)
ax3.set_xlabel("x")
ax3.set_ylabel("y")
ax3.set_title("z constant")

ax4.contourf(X, Y, Z)
ax4.set_xlabel("x")
ax4.set_ylabel("y")
ax4.set_title("z constant")
```
