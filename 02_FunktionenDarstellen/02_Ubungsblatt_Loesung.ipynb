### Aufgabe 1

Betrachten wir mal die Funktion

$$ g(x, y) = x \cdot e^{- x^2 - y^2} + 0.2y$$
<!-- $$ g(x, y) = \left(x^2 + 1.5 y^3\right) e^{- x^2 - y^2} $$ -->

im Wertebereich

$$x, y \in[-2,2]$$

**1.** Stelle die Höhenlinien dieser Funktion dar

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Aufgabe 1.1

n = 500
x = np.linspace(-2, 2, n)
y = np.linspace(-2, 2, n)
X, Y = np.meshgrid(x, y)
Z = X * np.exp(-X**2 - Y**2) + 0.2 * Y
plt.contour(X, Y, Z, cmap="coolwarm")
plt.xlabel("x")
plt.ylabel("y")
plt.colorbar(label="z")
plt.show()
```

**2.** Stelle Isokonturlinien für a) konstante $x$ und b) konstante $y$ dar (mit korrekten Achsenbeschriftungen).

```python
# Aufgabe 1.2 a):

cs = plt.contour(Y, Z, X)
cs.clabel(inline=True, fontsize=12)
plt.xlabel("y")
plt.ylabel("z")
plt.colorbar(label="x")
plt.show()
```

```python
# Aufgabe 1.2 b):

cs = plt.contour(X, Z, Y)
cs.clabel(inline=True, fontsize=10)
plt.xlabel("x")
plt.ylabel("z")
plt.colorbar(label="y")
plt.show()
```

**3.** Erstelle einen Höhenliniendiagramm mit gefüllten Flächen.

Nutze dazu die Funktion [`matplotlib.pyplot.contourf`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.contourf.html).

```python
# Aufgabe 1.3

plt.contourf(X, Y, Z, levels=20, cmap="coolwarm")
plt.colorbar(label="z")
plt.xlabel("x")
plt.ylabel("y")
plt.contour(X, Y, Z, levels=20, colors="black", linewidths=0.8)
plt.show()
```

**4.** Zeige alle Graphen in einem Plot

Benutze [`matplotlib.pyplot.subplots`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html), um eine Diagramm mit zwei mal zwei plots zu erzeugen, und füge dann die Contour Plots (wie oben) ein. Schreibe die Axen auch an.

*Optional:* Platziere die Plots so, dass die Achsen der Subplots jeweil horizontal und vertikal miteinander übereinstimmen. D.h. wenn die Höhenlinien (konst. $z$) unten links dargestellt sind, dann sollen darüber (oben links) die Linien konstanter $y$ geplottet werden, so dass die beiden horizontalen Achsen jeweils die Variable $x$ darstellen.

```python
# Create the contour plot
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(8, 8))

# Constant z contours on the lower left
cs = ax3.contour(X, Y, Z)
cs.clabel(inline=True, fontsize=9)
ax3.set_title("constant z")
ax3.set_xlabel("x")
ax3.set_ylabel("y")

# Constant x contours on the lower right
cs = ax4.contour(Z, Y, X)
cs.clabel(inline=True, fontsize=9)
ax4.set_title("constant x")
ax4.set_xlabel("z")
# ax4.set_ylabel('y')

# Constant y contours on the upper left
cs = ax1.contour(X, Z, Y)
cs.clabel(inline=True, fontsize=9)
ax1.set_title("constant y")
# ax1.set_xlabel('x')
ax1.set_ylabel("z")

# Empty subplot on the upper right
ax2.set_axis_off()
plt.show()
```

### Aufgabe 2:

**1.** Lies die Daten in von Daten_Woche2.csv ein

Die Daten sind Messwerte, die zu $x$- und $y$-Werten im Bereich $x,y\in[-2,2]$ gehören.


```python
# Aufgabe 2.1

Z = np.genfromtxt("Daten_Woche2.csv", delimiter=",")
Z.shape
```

**2.** Erstelle für die eingelesenen Daten alle Darstellungen wie in Aufgabe 1.

Nutze [`numpy.linspace`](https://numpy.org/doc/stable/reference/generated/numpy.linspace.html) und [`numpy.meshgrid`](https://numpy.org/doc/stable/reference/generated/numpy.meshgrid.html), um die zugehören $x$- und $y$-Matrizen zu erzeugen. Plotte die Contours und packe alle Plots in ein einziges Diagramm wie in Aufgabe 1.

```python
# Aufgabe 2.2

x = np.linspace(-2, 2, Z.shape[0])
y = np.linspace(-2, 2, Z.shape[1])
X, Y = np.meshgrid(x, y)

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(8, 8), constrained_layout=True)

ax1.contour(Y, Z, X)
ax1.set_xlabel("y")
ax1.set_ylabel("z")
ax1.set_title("x constant")

ax2.contour(X, Z, Y)
ax2.set_xlabel("x")
ax2.set_ylabel("z")
ax2.set_title("y constant")

ax3.contour(X, Y, Z)
ax3.set_xlabel("x")
ax3.set_ylabel("y")
ax3.set_title("z constant")

ax4.contourf(X, Y, Z)
ax4.set_xlabel("x")
ax4.set_ylabel("y")
ax4.set_title("z constant")

plt.show()
```
